```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
datos = pd.read_csv('data/data.csv',index_col=0)
```

```python
pd.set_option('display.max_columns', 32) 
datos.head()
```

```python
datos.describe()
```

```python
datos = datos.drop(['Unnamed: 32'], axis = 1) # Esta columna vienen toda con NaNs
datos.info()
```

```python
datos['diagnosis'] = datos['diagnosis'].map({'M':1,'B':0}) # Se codifica eel tipo de c치ncer de manera binaria
```

```python

from sklearn import preprocessing
datos_escalados = pd.DataFrame(preprocessing.scale(datos.iloc[:,1:32])) #Se dej칩 por fuera columna 'diagnosis'
datos_escalados.columns = list(datos.iloc[:,1:32].columns)
```

```python
datos_escalados.describe()
```

## K- Means

Lo primero que haremos en correr el agoritmo para diferentes n칰meros de clusters:

```python
from sklearn.cluster import KMeans
rango_clusters = range(1,15)
error_clusters = []

for num_clusters in rango_clusters:
    clusters = KMeans(num_clusters)
    clusters.fit(datos_escalados)
    error_clusters.append(clusters.inertia_)
    
clusters_df = pd.DataFrame({"num_clusters":rango_clusters,"error_clusters":error_clusters})
plt.plot(clusters_df.num_clusters,clusters_df.error_clusters,marker ="o")
plt.xlabel('N칰mero Clusters')
plt.ylabel('Error')
```

A la hora de decidir el n칰mero de clusters conveniente para los datos, es conveniente usar la -Elbow Rule-. Esta regla consiste en correr el algor칤tmo para varios valores de k, y graficar estos valores contra los errores.
Se usa en la pr치ctica un valor de k cercano al punto de 'cambio de la convexidad'. 

V칠ase:  https://en.wikipedia.org/wiki/Determining_the_number_of_clusters_in_a_data_set

En este caso, vamos a considerar el valor k=2, ya que adem치s de ser un buen valor serg칰n el criterio mencionado, nos va a servir para analizar como el tipo de cancer se ve relejado o no en los 2 clusters por crear.

```python
# Creamos el modelo con k = 2
# Cluster usando k- means
kmeans1 = KMeans(n_clusters=2, init='k-means++', n_init=10, max_iter=300, tol=0.0001, precompute_distances='auto', verbose=0, random_state=None, copy_x=True, n_jobs=1, algorithm='auto')
kY = kmeans1.fit_predict(datos_escalados)
```

## Visualizaci칩n

Una de las dificultades al crear algor칤tmos de clustering es la visualizaci칩n de los datos de altas dimensiones.

En la clase sklearn.manifold, tenemos el m칠todo llamado t-distributed Stochastic Neighbor Embedding, identificado como TSNE. Esta es una t칠cnica probabil칤stica que mantiene las estructuras locales de los datos en el espacio de incrustaci칩n de dimensi칩n menor. Nos genera una 'proyecci칩n' de los datos en en plano 2-dimensional.
V칠ase:


http://scikit-learn.org/stable/modules/generated/sklearn.manifold.TSNE.html


https://www.quora.com/How-does-t-SNE-work-in-simple-words

```python
# Se crea incrustaci칩n de los datos en dos dimensiones
from sklearn.manifold import TSNE
tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter= 4000, random_state = 2)
Y = tsne.fit_transform(datos_escalados)
```

```python
# Se grafica clusters modelados vs los reales seg칰n variable -diagnostic-

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 8))

ax1.scatter(Y[:,0],Y[:,1],  c=kY, cmap = "jet", edgecolor = "None", alpha=0.35)
ax1.set_title('Clusters por k - means')

ax2.scatter(Y[:,0],Y[:,1],  c = datos['diagnosis'], cmap = "jet", edgecolor = "None", alpha=0.35)
ax2.set_title('Clusters Actuales')
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report
print(confusion_matrix(datos['diagnosis'],kmeans1.labels_))
print(classification_report(datos['diagnosis'],kmeans1.labels_))
```

De esto, se puede decir que los clusters realmente capturan el estado del cancer de las personas, pues los indicadores son altos. Note que se cometi칩 m치s ruido al incluir pacientes con cancer maligno en benigno que al contrario.

# BIRCH (balanced iterative reducing and clustering using hierarchie


Es un algoritmo no supervisado de minado de datos usado para crear cluster jerarquicos particularmente sobre datasets muy grandes
Es un algoritmo de memoria eficiente, aprendizaje en linea como alternativa a MiniBatchKMeans.

El algoritmo construye un estructura de arboles con los centroides del cluster leidos desde la hoja. Estos pueden ser los centroides finales o pueden ser entrada para otro algoritmo de cluster como AgglomerativeClustering.



```python
from sklearn.cluster import Birch
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, perplexity=40, n_iter= 4000, random_state = 2)
Y = tsne.fit_transform(datos_escalados)

birch1 = Birch(n_clusters=2)
bY = birch1.fit_predict(datos_escalados)
bY = [0 if x==1 else 1 for x in bY]
```

```python
# Se grafica clusters modelados vs los reales seg칰n variable -diagnostic-

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(12, 8))

ax1.scatter(Y[:,0],Y[:,1],  c=bY, cmap = "jet", edgecolor = "None", alpha=0.35)
ax1.set_title('Clusters por Birch')

ax2.scatter(Y[:,0],Y[:,1],  c = datos['diagnosis'], cmap = "jet", edgecolor = "None", alpha=0.35)
ax2.set_title('Clusters Actuales')

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(datos['diagnosis'], bY))
print(classification_report(datos['diagnosis'], bY))
```

```python
birch1 = Birch(n_clusters=2, branching_factor=1000, threshold=0.8,)
bY = birch1.fit_predict(datos_escalados)
bY = [0 if x==1 else 1 for x in bY]
print(confusion_matrix(datos['diagnosis'], bY))
print(classification_report(datos['diagnosis'], bY))
```
